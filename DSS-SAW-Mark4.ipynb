```python
df5 = df.copy()
for index, row in df5.iterrows():
    for i in range(1, 11):
        # calculate weighted average
        row['P'+str(i)] = np.average(row['P'+str(i)], weights=[4,3,2,1])

# normalize for benefit 
for i in (4,6,7,8,9):
    arr = []
    for j in range (0,len(df)):
        arr.append(df5.iloc[j,i]) 
         
    maxx = max(arr)
    minn = min(arr)

    
    for j in range (0,len(df)):
        # print(df3.iloc[j,i])
        df5.iloc[j,i] = round(minn/df5.iloc[j,i],2)

# drop unused columns
df5 = df5.drop(['P1', 'P2','P3','P5','P10'], axis=1)

# criteria weight advised by an expert
criteria_weight = [3,4,1,2,5]

# calculate values times preference weight
for i in range (0,len(df)):
    for j in range (1,6):
        df5.iloc[i,j] = df5.iloc[i,j] * criteria_weight[j-1]

# sum all p1-p10 values
for i in range (0,len(df)):
    # print(df5.iloc[i,1])
    df5.iloc[i,1] = df5.iloc[i,1] + df5.iloc[i,2] + df5.iloc[i,3] + df5.iloc[i,4] + df5.iloc[i,5] 

#remove p2-p10
df5.drop(columns=df5.columns[2:11], inplace=True)

#rename p1 to SAW
df5.rename(columns={'P4': 'Teaching Skill'}, inplace=True)

df5.head()
```
