```python
# preprocessing, change all data to array of int
import statistics as st

for index, row in df.iterrows():
        for i in range(1, 11):
                # remove the first and last character
                row['P'+str(i)] = row['P'+str(i)][1:-1]

                #remove blankspace
                row['P'+str(i)] = row['P'+str(i)].replace(" ", "")

                # separate each value based on comma and put it in array, cast as int
                rowvalues = row['P'+str(i)].split(',')

                # cast as int
                rowvalues = list(map(int, rowvalues))
                row['P'+str(i)] = rowvalues
  
df.head()
```

```python
# EXP1
# weighted average calculation
df3 = df.copy()

for index, row in df3.iterrows():
    for i in range(1, 11):
        # calculate weighted average
        row['P'+str(i)] = np.average(row['P'+str(i)], weights=[4,3,2,1])
        
df3.head()
```

```python
# EXP1
# normalize for benefit 
for i in (1,2,3,5,10):
    arr = []
    for j in range (0,len(df)):
        arr.append(df3.iloc[j,i]) 
         
    maxx = max(arr)
    minn = min(arr)

    
    for j in range (0,len(df)):
        # print(df3.iloc[j,i])
        df3.iloc[j,i] = round(minn/df3.iloc[j,i],2)


# normalize for cost
for i in (4,6,7,8,9):
    arr = []
    for j in range (0,len(df)):
        arr.append(df3.iloc[j,i]) 
         
    maxx = max(arr)
    minn = min(arr)

    
    for j in range (0,len(df)):
        # print(df3.iloc[j,i])
        df3.iloc[j,i] = round(df3.iloc[j,i]/maxx,2)

df3.head()
```
