# Decision Support System - Simple Additive Weighting (SAW)

Simple Additive Weighting (SAW) is one method used to solve the problem of multi-attribute decision making. The basic concept SAW method is to find the sum of the weighted performance rating for each alternative on all attributes . SAW method requires a process of normalizing the decision matrix (X) to a scale that can be compared with all the ratings of existing alternatives.

$r_{ij} = \dfrac{x_{ij}}{Max(x_{ij})}$

$r_{ij} = \dfrac{x_{ij}}{Min(x_{ij})}$

If j is an attribute benefit then using the formula number one. If the attribute j cost then using the formula number two:

$w=\dfrac{C1}{C1 + ... + Cn}x100%$

$V_{i}=\sum_{j-1}^{n} ? 1 w_{j} r_{ij}$

The weights of all criteria are obtained by using the formula number three. With rij is the normalized performance rating of alternatives on attribute Ci Ai; i = 1,2, ..., n and j = 1,2, ..., n. Preference value alternative (vi) using the formula number four.
