```python
# preprocessing, change all data to array of int
import statistics as st

for index, row in df.iterrows():
        for i in range(1, 11):
                # remove the first and last character
                row['P'+str(i)] = row['P'+str(i)][1:-1]

                #remove blankspace
                row['P'+str(i)] = row['P'+str(i)].replace(" ", "")

                # separate each value based on comma and put it in array, cast as int
                rowvalues = row['P'+str(i)].split(',')

                # cast as int
                rowvalues = list(map(int, rowvalues))
                row['P'+str(i)] = rowvalues
  
df.head()
```
